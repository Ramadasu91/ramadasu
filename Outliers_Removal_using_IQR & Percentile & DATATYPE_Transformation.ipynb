```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('googleplaystore.csv')
```

```python
df.head()
```

```python
sns.distplot(df['Rating'])
```

```python
q1=df['Rating'].quantile(0.25)
q3=df['Rating'].quantile(0.75)
IQR=q3-q1
```

```python
Min=q1-1.5*IQR
Max=q3+1.5*IQR
```

```python
df[(df['Rating']<Min) | (df['Rating']>Max)]
```

### Trimming

```python
df[(df['Rating']>Min) & (df['Rating']<Max)]
```

### Capping

df['Rating']=np.where(df['Rating']>Max,Max,np.where(df['Rating']>Min,Min,df['Rating']))

## Outlier Removal using Percentile

```python
q1=df['Rating'].quantile(0.05)
q3=df['Rating'].quantile(0.95)
IQR=q3-q1
```

```python
Min=q1-1.5*IQR
Max=q3+1.5*IQR
```

```python
df[(df['Rating']<Min) | (df['Rating']>Max)]
```

### Trimming

```python
df[(df['Rating']>Min) & (df['Rating']<Max)]
```

### Capping

```python
df['Rating']=np.where(df['Rating']>Max,Max,np.where(df['Rating']>Min,Min,df['Rating']))
```
