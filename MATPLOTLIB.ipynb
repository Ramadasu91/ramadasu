```python
import matplotlib.pyplot as plt
import numpy as np
```

GENERATING SIN AND COS GRAPHS

```python
x=np.linspace(1,15,10)
y=np.sin(x)
z=np.cos(x)

plt.plot(x,y,label='sine wave',marker='*')
plt.plot(x,z,label='cosine wave',marker='_')

plt.title('Sine and Cosine Wave')
plt.legend()
plt.show()
```

<h1>SCATTER PLOT</h1>

```python
x=np.random.rand(100)
y=np.random.rand(100)
plt.scatter(x,y,marker='o',edgecolors='black',s=np.random.rand(100)*1000,c=np.random.rand(100),alpha=.8)
plt.colorbar()
```

<h1>HISTOGRAM</h1>

```python
plt.style.use('grayscale')
x=np.random.randn(1000000)
plt.hist(x,color='blue',edgecolor='white',bins=20000,alpha=.7)
```

```python
plt.style.use('seaborn-v0_8')
x=np.random.randn(1000)
plt.hist(x,color='red',edgecolor='white',bins=20,alpha=.7,histtype='stepfilled')
```

```python
x=np.random.randn(1000)
y=np.random.randn(1000)
plt.hist2d(x,y,cmap='viridis',bins=20)
```

```python
x=np.random.randn(1000)
y=np.random.randn(1000)+2
z=np.random.randn(1000)+4
plt.style.use('Solarize_Light2')

plt.hist(x,bins=10)
plt.hist(y,bins=10)
plt.hist(z,bins=10)
```
