```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('googleplaystore.csv')
```

```python
df.head()
```

## Univariate-Analysis Non-Graphical

### Categorical

```python
df['Category'].value_counts()
```

```python
df['Category'].value_counts(normalize=True)
```

```python
df['Type'].value_counts().reset_index()
```

### Numerical

```python
df['Rating'].value_counts()
```

```python
bins=(0,1,2,3,4,5)
```

```python
df['Rating'].value_counts(bins=bins)
```

```python
df['Rating'].skew()
```

```python
df['Rating'].kurt()
```

## UNivariate-Analysis Graphical

### Categorical

#### Barplot

```python
sns.countplot(data=df,x='Category')
```

```python
sns.countplot(data=df,x='Type')
```

```python
sns.countplot(data=df,x='Type',hue='Content Rating')
```

#### Piechart

```python
df['Category'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
df['Type'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
df['Content Rating'].value_counts().plot(kind='pie',autopct='%.2f')
```

### Numerical

#### Histogram

```python
sns.histplot(data=df[df['Rating']<7.5],x='Rating')
```

#### Distribution plot

```python
sns.distplot(df['Rating'])
```

#### Boxplot

```python
sns.boxplot(df['Rating'])
```

#### Voilinplot

```python
sns.violinplot(df['Rating'])
```
