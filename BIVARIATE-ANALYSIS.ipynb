```python
import pandas as pd
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df=sns.load_dataset("titanic")
df.head()
```

### Bivariate-Analysis Categorical-Numerical

#### Barplot

```python
sns.barplot(x='class',y='age',data=df)
```

```python
sns.barplot(x='class',y='age',data=df,estimator='sum')
```

```python
sns.barplot(x='class',y='age',hue='sex',data=df)
```

#### Boxplot

```python
sns.boxplot(x='class',y='age',data=df)
```

```python
sns.boxplot(x='class',y='age',hue='survived',data=df)
```

#### violinplot

```python
sns.violinplot(x='class',y='age',data=df)
```

## Categorical

#### Countplot

```python
sns.countplot(x='survived',hue='class',data=df)
```

#### Heatmap

```python
x=pd.crosstab(df['survived'],df['class'])
x
```

```python
sns.heatmap(x,annot=True,fmt='.1f')
```

```python
sns.heatmap(x,annot=True,fmt='.1f',cmap='coolwarm')
```

#### Barplot

```python
sns.barplot(x='survived',y='age',hue='sex',data=df)
```

```python
sns.barplot(x='class',y='age',hue='sex',data=df)
```

#### Stripplot

```python
sns.stripplot(x='survived',y='age',hue='sex',data=df)
```

#### Pairplot

```python
df2=sns.load_dataset("iris")
df2.head()
```

```python
df2.head(10).plot(kind='bar',stacked=True)
```

```python
sns.pairplot(df2)
```

## Numeric

#### scatter

```python
sns.scatterplot(x='age',y='fare',data=df)
```

```python
sns.scatterplot(x='age',y='fare',hue='survived',data=df)
```

```python
sns.scatterplot(x='age',y='fare',hue='survived',style='sex',data=df)
```

#### Jointplot

```python
df3=sns.load_dataset('tips')
df3.head()
```

```python
sns.jointplot(x='total_bill',y='tip',data=df3)
```

```python
sns.jointplot(x='total_bill',y='tip',hue='sex',data=df3)
```

```python
sns.jointplot(x='total_bill',y='tip',kind='reg',data=df3)
```
